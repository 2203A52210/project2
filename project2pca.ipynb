```python
import pandas as pd
a=pd.read_csv("/content/HepatitisCdata.csv")
print(a)
```

```python
y=a['Category']
print(y)
```

```python
x=a.drop('Category',axis=1)
x
```

```python
a.drop(labels='Unnamed: 0',axis=1,inplace=True)
a
```

```python
x['Sex'] = x['Sex'].replace({'m': 1, 'f' : 0})
x
```

```python
y = y.replace({'0=Blood Donor':0,'0s=suspect Blood Donor':1,'1=Hepatitis':2,'2=Fibrosis':3,'3=Cirrhosis':4})
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
model=SVC(kernel='linear')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
n_components = 2
pca = PCA(n_components=n_components)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train_pca, y_train)
accuracy = logistic_model.score(x_test_pca, y_test)
print(f"Accuracy of logistic regression after PCA: {accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.utils import resample
n_iterations = 1000
bootstrap_accuracies = []
for _ in range(n_iterations):
    x_train_boot, y_train_boot = resample(x_train, y_train)
    logistic_model.fit(x_train_boot, y_train_boot)
    accuracy = logistic_model.score(x_test, y_test)
    bootstrap_accuracies.append(accuracy)
plt.figure(figsize=(8, 6))
plt.hist(bootstrap_accuracies, bins=30, color='skyblue')
plt.title('Distribution of Accuracies (Bootstrap)')
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.show()
```
